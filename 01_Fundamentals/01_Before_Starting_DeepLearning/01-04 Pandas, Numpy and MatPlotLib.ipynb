# Pandas

```python
import pandas as pd
```

```python
sr = pd.Series([17000, 18000, 1000, 5000],
               index=["피자", "치킨", "콜라", "맥주"])
print("시리즈 출력 : ")
print('-'*15)
print(sr)
```

```python
print('시리즈의 값 : {}'.format(sr.values))
print('시리즈의 인덱스 : {}'.format(sr.index))
```

```python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df = pd.DataFrame(values, index=index, columns=columns)

print('데이터프레임 출력 :')
print('-'*18)
print(df)
```

```python
print('데이터프레임의 인덱스 : {}'.format(df.index))
print('데이터프레임의 열이름 : {}'.format(df.columns))
print('데이터프레임의 값 : ')
print('-'*18)
print(df.values)
```

# Numpy

```python
import numpy as np
```

```python
vec = np.array([1, 2, 3, 4, 5])
print(vec)
```

```python
mat = np.array([[10, 20, 30], [60, 70, 80]])
print(mat)
```

```python
print('vec의 타입 :', type(vec))
print('mat의 타입 :', type(mat))
```

```python
print('vec의 축의 개수 : ', vec.ndim)
print('vec의 크기(shape) : ', vec.shape)
```

```python
print('mat의 축의 개수 : ', mat.ndim)
print('mat의 크기(shape) : ', mat.shape)
```

```python
zero_mat = np.zeros((2,3))
print(zero_mat)
```

```python
one_mat = np.ones((2,3))
print(one_mat)
```

```python
same_value_mat = np.full((2,2), 7)
print(same_value_mat)
```

```python
eye_mat = np.eye(3)
print(eye_mat)
```

```python
random_mat = np.random.random((2,2))
print(random_mat)
```

```python
range_vec = np.arange(10)
print(range_vec)
```

```python
n = 2
range_n_step_vec = np.arange(1, 10, n)
print(range_n_step_vec)
```

```python
reshape_mat = np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
```

```python
mat = np.array([[1,2,3], [4, 5, 6]])
print(mat)
```

```python
slicing_mat = mat[0, :]
print(slicing_mat)
```

```python
slicing_mat = mat[:, 1]
print(slicing_mat)
```

```python
mat = np.array([[1,2], [4, 5], [7,8]])
print(mat)
```

```python
print(mat[1,0])
```

```python
indexing_mat = mat[[2, 1], [0, 1]]  #익숙하거나 자연스럽진 못하네
print(indexing_mat)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

result = x + y
print(result)
```

```python
result = x - y
print(result)
```

```python
result = result * x
print(result)
```

```python
result = result / x
print(result)
```

```python
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
mat3 = np.dot(mat1, mat2)
print(mat3)
```
