###Implementing adder with a function


```python
result = 0

def add(num):
  global result
  result += num
  return result
```

```python
print(add(3))
print(add(4))
```

###Implementing adder with a class

```python
class Calculator:
  def __init__(self):
    self.result = 0;

  def add(self, num):
    self.result += num
    return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```
